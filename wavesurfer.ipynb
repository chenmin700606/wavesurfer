```python
import torchaudio
from wavesurfer import wavesurfer

wav = "data/test_16k.wav"
waveform, sr = torchaudio.load(wav)
wavesurfer.display_audio(waveform, sr, width=800, enable_spectrogram=True)

waveform, sr = torchaudio.load(wav, frame_offset=1 * 16000, num_frames=2 * 16000)
wavesurfer.display_audio(waveform, sr, width=1200, enable_minimap=True)
```
